```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
#loading the given file monet.csv into dataset

dataset = pd.read_csv('monet.csv')

dataset
```

```python
#Task -1:
#model-1 
# Creatig new independent variable as size = height * width
#it is important to determine cost

height = dataset['HEIGHT']
width = dataset['WIDTH']
```

```python
dataset['SIZE'] = height * width
dataset['SIZE']

# considering size = height * width
```

```python
# determining the independent variable SIZE as x and dependent variable PRICE as y
# Considering one variable as PRICE and the other as SIZE = height * width

x = dataset['SIZE'].values
y = dataset['PRICE'].values

#splitting the dataset
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.5,random_state=40)
```

```python
x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python

regr=LinearRegression()
```

```python
regr1=regr.fit(x_train,y_train)
#this is fitting the simple regression to linear regression
```

```python
y_predic=regr.predict(x_test)
y_predic

#The above is for predicting the test results
```

```python
#mean square error
np.sqrt(mean_squared_error(y_test,y_predic))
```

```python
#Scatter plot with regression line

plt.scatter(x,y,color='blue')  
#this is actual data
plt.plot(x_train,regr.predict(x_train),color='red')
plt.title('Size vs Price ')
plt.xlabel('SIZE')
plt.ylabel('PRICE')

#the above is for visualizing the size 
```
